```python
def get_file(ee,model,field,yr):
    if model == 'CESM1':
        topdir = ('/glade/collections/cdg/data/cesmLE/CESM-CAM5-BGC-LE/'+
                  'lnd/proc/tseries/monthly/')
        matchme = 'b.e11.B20TRC5CNBDRD.f09_g16.'
        lenskey = list(1+np.arange(35))+list(101+np.arange(7))
        nn      = lenskey[ee]
        ff = glob.glob(topdir+field+'/'+matchme+str(nn).zfill(3)+'*.nc')[0]
    else:
        topdir = '/glade/campaign/cesm/collections/CESM2-LE/archive/'
        mlist = sorted(glob.glob(topdir+'*BHISTcmip6.f09_g17*'))
        mm = mlist[ee]
        lnd    = '/lnd/proc/tseries/month_1/'
        firstyear = 10*int(yr/10)
        yy    = '.'+str(firstyear)+'01-'+str(firstyear+9)+'12.'
        if yr>2009:
            yy = '.201001-201412.'
        ff     = glob.glob(mm+lnd+'*.h0.'+field+yy+'*')[0]
    return ff
```

```python
def all_files(model,field,firstyear):
    if model == 'CESM1':
        nens = 42
        files = [get_file(ee,'CESM1',field,0) for ee in range(nens)]
    else:
        nens = 50
        yrs = [*np.arange(firstyear,2009,10),2010]
        files = [[get_file(ee,'CESM2',field,yr) for yr in yrs] 
                 for ee in range(nens)]
    return files
```

```python
def get_monthly(field,model,firstyear):
    
    yr1 = {'CESM1':1920,'CESM2':1850}
    
    firstyear = max(yr1[model],firstyear)
    dsets = []
    first = True
   
    for field in fields:
        files = all_files(model,field,firstyear)
        
        if first:
            ensdim  = xr.DataArray(np.arange(len(files)), dims='ens', name='ens')
            ensdim.attrs['long_name']='ensemble number'
            concat_dim = {'CESM1':ensdim,'CESM2':[ensdim,'time']}

        ds = xr.open_mfdataset(files,combine='nested',parallel=True,
                               concat_dim=concat_dim[model])
        nmonths = len(ds.time)
        yr0 = ds['time.year'][0].values
        ds['time'] =xr.cftime_range(str(yr0),periods=nmonths,freq='MS')
        
        if first: 
            landarea = (ds['area']*ds['landfrac']).sel(ens=0)
            if 'time' in landarea.coords:
                landarea = landarea.isel(time=0)
            landarea.name = 'landarea'
            landarea.attrs['units'] = 'km2'
            dsets.append(landarea)
            first = False
                 
        ix = ds['time.year']>=firstyear
        with dask.config.set(**{'array.slicing.split_large_chunks': True}):
            ds = ds.isel(time=ix)
        dsets.append(ds[field])
     
    return dsets
```

```python
def month_wts(nyears):
    days_pm  = [31,28,31,30,31,30,31,31,30,31,30,31]
    return xr.DataArray(np.tile(days_pm,nyears),dims='time')
```

```python
%%time
model = 'CESM1'
fields = ['QVEGT','GPP']     #needs brackets, even for single var 
firstyear = 1850
landarea,qvegt,gpp = get_monthly(fields,model,firstyear)
```

### Example 1: plot global annual gpp

```python
nyears  = len(np.unique(gpp['time.year']))
cf = 24*60*60           # used to convert /s to /d
dpm = month_wts(nyears) # days per month
gpp_ann = cf*(dpm*gpp).groupby('time.year').sum()
qv_ann  = cf*(dpm*qvegt).groupby('time.year').sum()
```

```python
cf1 = 1e6    #km2-->m2
cf2 = 1e-15  #gC-->PgC and kgH2O-->TtH2O
gpp_ann_glob = cf1*cf2*(landarea*gpp_ann).sum(dim=['lat','lon'])
```

```python
%%time
##  force the calculation:
##     xarray uses lazy evalutation, avoiding crunching the numbers 
##     up to this point, I haven't actually done any calculations!
##     xarray waits until I explicitly required to do so
##     I try to avoid having that occur in the plotting call
##  Note that if you plan on reusing an intermediate result:
##     e.g. annual averages, 
##     you should probably force the calc and save it at that step
gpp_ann_glob = gpp_ann_glob.compute()
```

```python
t = np.unique(gpp['time.year'])
plt.plot(t,gpp_ann_glob,'.',color=[0.6,0.6,0.6])
plt.xlabel('year')
plt.ylabel('GPP (PgC)')
plt.title('CESM1-LE (42 members)');
```

### Example 2: plot global annual WUE over time

```python
nyears  = len(np.unique(gpp['time.year']))
cf = 24*60*60           # used to convert /s to /d
dpm = month_wts(nyears) # days per month
cf1 = 1e6    #km2-->m2
cf2 = 1e-15  #gC-->PgC and kgH2O-->TtH2O
```

```python
%%time
gpp_ann = cf*(dpm*gpp).groupby('time.year').sum()
gpp_ann_glob = cf1*cf2*(landarea*gpp_ann).sum(dim=['lat','lon']).compute()
qv_ann  = cf*(dpm*qvegt).groupby('time.year').sum()
qv_ann_glob = cf1*cf2*(landarea*qv_ann).sum(dim=['lat','lon']).compute()
```

```python
%%time
wue_glob=(gpp_ann_glob/qv_ann_glob).compute()
```

```python
t = np.unique(gpp['time.year'])
plt.plot(t,wue_glob,'.',color=[0.6,0.6,0.6]);
plt.ylabel('Global Annual WUE (gC/kgH2O)')
plt.xlabel('year')
plt.title('CESM1-LE');
```
